Conduct an unpaired t-test assuming equal variances
* shore vs dock
* dock vs pipe
* shore vs pipe


```python


# Data
shore = [-0.024868123587, 0.01582517, 0.047475509, -0.0535041446873]
dock = [-0.0233609645818, -0.06857573474, 1.019593067, -0.0535041446873]
pipe = [0.047475509, 0.008289375, 0.389600603, 0]

# Unpaired t-test assuming equal variances
t_stat_shore_dock, p_value_shore_dock = stats.ttest_ind(shore, dock, equal_var=True)
t_stat_dock_pipe, p_value_dock_pipe = stats.ttest_ind(dock, pipe, equal_var=True)
t_stat_shore_pipe, p_value_shore_pipe = stats.ttest_ind(shore, pipe, equal_var=True)

print(f"Shore vs Dock: t-statistic = {t_stat_shore_dock}, p-value = {p_value_shore_dock}")
print(f"Dock vs Pipe: t-statistic = {t_stat_dock_pipe}, p-value = {p_value_dock_pipe}")
print(f"Shore vs Pipe: t-statistic = {t_stat_shore_pipe}, p-value = {p_value_shore_pipe}")
```
